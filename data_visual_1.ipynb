```python
#code here 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#code here 
data = pd.read_csv('/content/Car_Purchasing_Data.csv',encoding="ISO-8859-1")
data #data.head() to show first 5 rows
```

```python
data.drop(['Customer Name','Customer e-mail','Country',],axis =1, inplace =True)
```

```python
#code here 

data.isnull().sum()
```

```python
data.dropna(subset = ['Credit Card Debt'],inplace =True) #'Credit Card Debt' lo null value so we gonna drop them
```

```python
#code here 
data['Car Purchase Amount'].fillna(data['Car Purchase Amount'].mean() ,inplace=True) #for Car Purchase Amount
```

```python
#code here 
data.isnull().sum()
```

```python
#code here 
sns.boxplot(data=data,palette='rocket',orient='h')    #we could show each graph separately to be accurate

sns.set(rc = {'figure.figsize':(20, 10 )})   #this function used to make the table more readable

#outlier =  [ x < (Q1 - 1.5IQR) , X > (Q3 + 1.5IQR) ]
```

```python
#code here 

sns.pairplot(data,hue ='Gender', palette ='mako')
```

```python
#code here   extra vesualization
sns.countplot(x='Gender',data =data , palette ='mako') #it appeares that both genders are near 
```
